This is the very first exercise in machine learning . The study case is about Iris flower. Iris flower has 3 types, which are Virginica, Versicolor, and Setosa. 
The purpose of this exercise is to classify the Iris flower based on its petal and sepal size. 
To do so, we can use multiclass classification and Decision Tree algorithm.
image.png

```python
import pandas as pd
from sklearn.datasets import load_iris

# Read the csv file
iris = pd.read_csv('Iris.csv')
```

```python
# display first five rows
iris.head()
```

```python
# drop unnecessary column
iris.drop('Id',axis=1,inplace=True)
```

```python
# separate the attribute and label
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']
```

```python
from sklearn.tree import DecisionTreeClassifier

# creating Decision Tree model
tree_model = DecisionTreeClassifier()

# training the model
tree_model.fit(X, y)
```

```python
# using model to predict iris flower
tree_model.predict([[6.2, 3.4, 5.4, 2.3]])
```

```python
# using graphviz to visualize the decision tree
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = 'iris_tree.dot',
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
    rounded = True,
    filled = True
)
```
